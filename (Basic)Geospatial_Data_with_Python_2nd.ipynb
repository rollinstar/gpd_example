```python
# 라이브러리 추가하기
import geopandas as gpd
import pandas as pd
```

```python
# 서울시 법정동 geojson 파일
_seoul_umd = './maps/seoul_umd.geojson'
```

```python
# GeoDataFrame
seoul_umd_gdf = gpd.read_file(_seoul_umd)
seoul_umd_gdf.head()
```

```python
# 법정동 지도 보기
seoul_umd_gdf.plot()
```

```python
# Dissolve를 이용해 시군구 데이터 만들기
seoul_sgg = seoul_umd_gdf.dissolve(by='COL_ADM_SE')
seoul_sgg
```

```python
seoul_sgg.reset_index(inplace=True)
seoul_sgg
```

```python
seoul_sgg.plot()
```

```python
# 서울시 인구 통계정보 가져오기 (Pandas)
seoul_sgg_pop = pd.read_excel('./files/seoul_sgg_pop.xlsx')
seoul_sgg_pop.head()
```

```python
seoul_sgg.info()
```

```python
seoul_sgg_pop.info()
```

```python
# 시군구 인구 파일 코드 문자열로 변환하기
seoul_sgg_pop = seoul_sgg_pop.astype({'SIG_CD':'str'})
seoul_sgg_pop.info()
```

```python
# 시군구 데이터에 인구데이터 병합하기
seoul_sgg = pd.merge(seoul_sgg, seoul_sgg_pop, how='left', left_on='COL_ADM_SE', right_on='SIG_CD')
seoul_sgg.head()
```

```python
# 시군구 geojson 생성하기
# 불필요한 컬럼 제거하기
columns = ['SIG_CD','SIG_KOR_NM','SEDAE','SEDAEDANGINGU','TOTAL_POP','MALE_POP','FEMALE_POP','65_OVER_AGE','geometry']
gdf = gpd.GeoDataFrame(seoul_sgg[columns], geometry='geometry')
gdf.to_file('./maps/seoul_sgg_final.geojson' , driver='GeoJSON')
```

# 스타벅스

```python
seoul_starbucks = pd.read_excel('./files/seoul_starbucks.xlsx')
seoul_starbucks.head()
```

```python
geometry = gpd.points_from_xy(seoul_starbucks.LONG, seoul_starbucks.LAT)
coulumns = ['STORE_NAME', 'STORE_TYPE', 'SIG_CD']
seoul_sb_gdf = gpd.GeoDataFrame(seoul_starbucks[coulumns], geometry=geometry)
seoul_sb_gdf.head()
```

```python
seoul_sb_gdf.to_file('./maps/seoul_starbucks.geojson', driver='GeoJSON')
```
